```python
def extract_audio_segment(audio_file_path, start_second, end_second, padding_before=0.5, padding_after=3.5):
    audio = AudioSegment.from_file(audio_file_path)
    start_ms = int((start_second - padding_before) * 1000)
    end_ms = int((end_second + padding_after) * 1000)
    return audio[start_ms:end_ms]

def save_audio_segment(audio_file_path, output_dir, start_second, end_second, padding_before=0.5, padding_after=3.5):
    audio_segment = extract_audio_segment(audio_file_path, start_second, end_second, padding_before, padding_after)
    output_file_path = os.path.join(output_dir, os.path.basename(audio_file_path).rsplit('.',1)[0])
    output_file_path = output_file_path.replace(' ', '_')
    output_file_path = output_file_path+'_'+str(start_second)
    audio_segment.export(output_file_path, format="wav")

def main(excel_file_path, audio_base_dir, output_base_dir):
    # Read the Excel file with multiple sheets
    xls = pd.ExcelFile(excel_file_path)

    # Process each sheet in the Excel file
    print(f'sheets: {xls.sheet_names}')
    for sheet_name in xls.sheet_names:
        # if sheet_name != 'BMR3 vs BMR5':
        #   continue
        print(f'starting sheet {sheet_name}')
        df = xls.parse(sheet_name)

        # Extract the first word of the sheet name and use it as the audio directory name
        audio_dir_name = sheet_name.split()[0]
        audio_dir = os.path.join(audio_base_dir, audio_dir_name)
        print(audio_dir_name)
        print(f'using {audio_dir}')

        # Create a directory for the current sheet (using the sheet name) in the output directory
        output_dir = os.path.join(output_base_dir, sheet_name)
        output_dir = output_dir.replace(' ','_')
        os.makedirs(output_dir, exist_ok=True)

        # Process each row in the DataFrame
        for index, row in df.iterrows():
            try:
              label = row['Head Drums']
            except Exception:
              continue
            start_second = row['time (sec)']

            filename = sheet_name + '.wav'
            audio_file_path = os.path.join(audio_dir, filename)
            if not os.path.isfile(audio_file_path):
                print(audio_file_path)
                print(f"File '{filename}' not found in the specified directory.")
                continue

            # Assuming end_second is the next row's time (sec) value
            next_index = index + 1
            if label == '*':
              save_audio_segment(audio_file_path, output_dir, start_second, start_second)

```

```python
excel_file_path = "/content/drive/MyDrive/Project/mole rats data/new HD and Vocal times.xlsx"
audio_directory = "/content/drive/MyDrive/Project/mole rats data/"
output_base_directory = "/content/drive/MyDrive/Project/data_longer_new"
main(excel_file_path, audio_directory, output_base_directory)
```
