```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

def detect_sound_pulse(audio_file, threshold=0.1):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)

    # Perform amplitude thresholding
    pulse_indices = np.where(y >= threshold)

    # Calculate time positions of detected pulses
    pulse_times = librosa.samples_to_time(pulse_indices, sr=sr)

    return y, pulse_times, sr

if __name__ == "__main__":
    # Provide the path to your WAV audio file
    audio_file_path = "/content/drive/MyDrive/Project/data/BMR2_vs_BMR3/BMR2_vs_BMR3_868"

    # Adjust the threshold value as needed (a value between 0 and 1)
    threshold_value = 0.1

    audio_signal, detected_pulses, sr = detect_sound_pulse(audio_file_path, threshold=threshold_value)

    # Create a time array for plotting the waveform
    time = np.arange(0, len(audio_signal)) / sr

    # Plot the waveform
    plt.figure(figsize=(12, 6))
    plt.plot(time, audio_signal, label='Audio Signal')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title('Audio Signal with Detected Pulses')
    plt.axhline(y=threshold_value, color='r', linestyle='--', label='Threshold')
    plt.scatter(detected_pulses, [threshold_value] * len(detected_pulses), c='g', marker='o', label='Detected Pulses')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

def audio_to_frequency_transform(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)

    # Compute the Short-Time Fourier Transform (STFT)
    stft = librosa.stft(y)

    # Convert the STFT magnitude to decibels
    stft_db = librosa.amplitude_to_db(np.abs(stft), ref=np.max)

    # Get the corresponding frequency and time axes
    frequency_bins = librosa.fft_frequencies(sr=sr, n_fft=stft.shape[0])
    time = librosa.frames_to_time(np.arange(stft.shape[1]), sr=sr)

    return stft_db, frequency_bins, time

if __name__ == "__main__":
    # Provide the path to your WAV audio file
    audio_file_path = "/content/drive/MyDrive/Project/data/BMR6_vs_BMR5829/BMR6_vs_BMR5829_776"

    stft_db, frequency_bins, time = audio_to_frequency_transform(audio_file_path)

    # Visualize the frequency Fourier transform
    plt.figure(figsize=(10, 6))
    plt.imshow(stft_db, cmap='viridis', origin='lower', aspect='auto',
               extent=[time.min(), time.max(), frequency_bins.min(), frequency_bins.max()])
    plt.colorbar(format='%+2.0f dB')
    plt.title('Frequency Fourier Transform (STFT)')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.tight_layout()
    plt.show()
```

```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

def audio_to_stft(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)

    # Compute the Short-Time Fourier Transform (STFT)
    _, _, Zxx = spectrogram(y, fs=sr, nperseg=512, noverlap=256)

    # Get the corresponding frequency and time axes
    frequency_bins = np.fft.fftfreq(Zxx.shape[0], d=1/sr)
    time = np.arange(0, Zxx.shape[1]) * (Zxx.shape[0] / sr)

    return Zxx, frequency_bins, time

def audio_to_frequency_transform(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file, sr=None)

    # Compute the Short-Time Fourier Transform (STFT)
    stft = librosa.stft(y)

    # Convert the STFT magnitude to decibels
    stft_db = librosa.amplitude_to_db(np.abs(stft), ref=np.max)

    # Get the corresponding frequency and time axes
    frequency_bins = librosa.fft_frequencies(sr=sr, n_fft=stft.shape[0])
    time = librosa.frames_to_time(np.arange(stft.shape[1]), sr=sr)

    return stft_db, frequency_bins, time

def save_stft_as_image(audio_file, output_dir):
    # Compute the STFT for the audio file
    stft, frequency_bins, time = audio_to_frequency_transform(audio_file)

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Extract the filename without extension from the audio file path
    filename = os.path.splitext(os.path.basename(audio_file))[0]

    # Save the STFT plot as an image
    plt.figure(figsize=(10, 6))
    plt.imshow(np.abs(stft), cmap='viridis', origin='lower', aspect='auto',
               extent=[time.min(), time.max(), frequency_bins.min(), frequency_bins.max()])
    plt.colorbar()
    plt.title('Short-Time Fourier Transform (STFT)')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.tight_layout()

    # Save the image with the same name as the audio file
    output_file = os.path.join(output_dir, f"{filename}_stft.png")
    plt.savefig(output_file)
    plt.close()

if __name__ == "__main__":
    # Provide the path to the directory containing WAV audio files
    audio_directory = "/content/drive/MyDrive/Project/data/BMR2_vs_BMR3"

    # Provide the path to the output directory where STFT images will be saved
    output_directory = "/content/drive/MyDrive/Project/data_stft2/BMR2_vs_BMR3"

    # Process each audio file in the directory
    for root, _, files in os.walk(audio_directory):
        for file in files:
            # if file.lower().endswith(".wav"):
                audio_file_path = os.path.join(root, file)
                save_stft_as_image(audio_file_path, output_directory)
```
